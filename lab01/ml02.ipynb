```python
import zipfile ##
import os

# Path to the downloaded zip file 
zip_file_path = r'C:\Users\Mahi2\projects\applied-ml-mk\applied-ml-mk\oral-cancer-prediction-dataset.zip'
extracted_folder = 'oral-cancer-dataset/'

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_folder)

print(f"Dataset extracted to {extracted_folder}")

import pandas as pd

# Update with the path where the dataset is located
oralcans1 = pd.read_csv(r'C:\Users\Mahi2\projects\applied-ml-mk\applied-ml-mk\lab01\oral-cancer-dataset\oral_cancer_prediction_dataset.csv')

oralcans1.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a countplot (categorical histogram) for 'Tobacco Use'
plt.figure(figsize=(8, 6))  # Set the figure size
sns.countplot(data=oralcans1, x='Tobacco Use', palette='viridis')

# Set labels and title
plt.title('Distribution of Tobacco Use')
plt.xlabel('Tobacco Use')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the histogram of 'Tobacco Use'
plt.figure(figsize=(8, 6))
sns.histplot(oralcans1['Tobacco Use'], bins=3, kde=False, color='brown')

# Add labels and title
plt.title('Histogram of Tobacco Use', fontsize=16)
plt.xlabel('Tobacco Use', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=oralcans1, x='Cancer Stage', hue='Tobacco Use', multiple="stack", kde=True, bins=30)

# Add labels and title
plt.title('Distribution of Cancer Stage by Tobacco Use', fontsize=16)
plt.xlabel('Cancer Stage', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

```python
# Create a histogram for Cancer Stage by Alcohol Consumption
plt.figure(figsize=(8, 6))
sns.histplot(data=oralcans1, x='Cancer Stage', hue='Alcohol Consumption', multiple="stack", kde=True, bins=30)

# Add labels and title
plt.title('Distribution of Cancer Stage by Alcohol Consumption', fontsize=16)
plt.xlabel('Cancer Stage', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot
plt.show()
```

```python
oralcans1['Lifestyle Risk Factor'] = oralcans1.apply(
    lambda row: 'Low Risk' if row['Tobacco Use'] == 'No' and row['Alcohol Consumption'] == 'No' else
                ('Moderate Risk' if row['Tobacco Use'] == 'No' and row['Alcohol Consumption'] == 'Yes' else
                 ('High Risk' if row['Tobacco Use'] == 'Yes' and row['Alcohol Consumption'] == 'No' else 'Very High Risk')),
    axis=1
)
print(oralcans1[['Tobacco Use', 'Alcohol Consumption', 'Lifestyle Risk Factor']].head())

# Check the distribution of the 'Lifestyle Risk Factor' to make sure the transformation worked as expected
print(oralcans1['Lifestyle Risk Factor'].value_counts())
```
